```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("./books.csv", delimiter=',')
```

```python
df.category.value_counts().plot.barh()
```

```python
df.category.unique()
```

```python
selected_categories = ['Medical','Art-Photograph','Biography','Childrens-Books','Computing','Home-Garden']
sub_df = pd.DataFrame(columns=df.columns)
samples_per_category = 20
sub_df
for cat in selected_categories:
    cat_df = df.loc[df.category.isin(selected_categories)][:samples_per_category]
    sub_df = pd.concat([sub_df,cat_df])
# sub_df.shape
sub_df
sub_df.shape
```

```python
# sub_df.to_csv("sub_books.csv")
sub_df.columns
```

```python
sub_df = sub_df.drop(['old_price','img_paths'],axis=1)
```

```python
sub_df = sub_df.rename(columns = {'name':'title', 'image':'imageLink'})
```

```python
sub_df.columns
```

```python
sub_df = sub_df.reset_index(drop=True)
```

```python
sub_df.to_csv('sub_kaggle_books.csv',index=False)
```
